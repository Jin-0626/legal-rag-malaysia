# Web Scraping Malaysian Legal Documents with BeautifulSoup4

This notebook demonstrates how to scrape legal documents from Malaysian portals using BeautifulSoup4

## Step 1: Import Required Modules

```python
import sys
from pathlib import Path

# Add src to path
sys.path.insert(0, str(Path.cwd().parent))

from src.web_scraper import LawScraper
from config.settings import RAW_DOCS_DIR

print("OK: Web scraper module imported")
```

## Step 2: Set Up Download Directory

```python
download_dir = RAW_DOCS_DIR / "downloaded"
download_dir.mkdir(exist_ok=True)

print(f"Download directory: {download_dir}")
print(f"Directory exists: {download_dir.exists()}")
```

## Step 3: Initialize Scraper

```python
# Initialize scraper with BeautifulSoup4
scraper = LawScraper(download_dir=str(download_dir))

print("OK: Scraper initialized with BeautifulSoup4")
print(f"Download directory: {scraper.download_dir}")
```

## Step 4: Scrape LOM Portal

```python
# Scrape LOM portal for Acts
print("Scraping LOM Portal for Malaysian Acts...\n")

lom_documents = scraper.scrape_lom_acts(
    url="https://lom.agc.gov.my/",
    max_documents=5  # Limit to 5 for testing
)

print(f"\nFound {len(lom_documents)} documents from LOM Portal")
for i, doc in enumerate(lom_documents, 1):
    print(f"\n{i}. {doc['title']}")
    print(f"   URL: {doc['url']}")
    print(f"   File: {doc['filename']}")
```

## Step 5: Download PDFs

```python
# Download the scraped PDFs
print("Downloading PDFs...\n")

downloaded_files = scraper.download_multiple_pdfs(lom_documents)

print(f"\nSuccessfully downloaded {len(downloaded_files)} files:")
for filepath in downloaded_files:
    print(f"  - {filepath}")
```

## Step 6: Verify Downloads

```python
# List downloaded files
pdf_files = list(download_dir.glob("*.pdf"))

print(f"PDFs in download directory ({download_dir}):")
for pdf in pdf_files:
    size_mb = pdf.stat().st_size / (1024 * 1024)
    print(f"  - {pdf.name} ({size_mb:.2f} MB)")

if not pdf_files:
    print("  No PDFs downloaded yet")
```
